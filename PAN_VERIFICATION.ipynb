```python
import pytesseract
from PIL import Image

def extract_name(image_path):
    """
    Extracts the name from an image using dynamic region detection.

    Args:
        image_path: The path to the image file.

    Returns:
        The extracted name as a string, or None if not found.
    """
    try:
        # Load the image using PIL
        img = Image.open(image_path)

        # Perform OCR with layout analysis
        data = pytesseract.image_to_data(img, output_type=pytesseract.Output.DICT)
        print(data)
        # Search for the keyword "Name" or similar keywords
        keywords = ["Name", "Father's Name"]
        name_index = -1  # Initialize with -1 to indicate not found
        for keyword in keywords:
            try:
                name_index = data['text'].index(keyword)
                break  # Exit loop if keyword is found
            except ValueError:
                pass  # Continue to next keyword if not found

        if name_index == -1:
            print("Name keyword not found in the image.")
            return None

        # Get the bounding box of the keyword
        x, y, w, h = data['left'][name_index], data['top'][name_index], data['width'][name_index], data['height'][name_index]
        print(x)
        print(y)
        print(w)
        print(h)
        # Define the region for the name field (adjust as needed)
        name_region = (x-75, y+h,  x + w + 300 , y+h+50)  # Adjust values based on image layout

        # Extract text from the name region
        name_text = pytesseract.image_to_string(img.crop(name_region), config='--psm 6')  # Adjust psm if needed

        # Clean and return the extracted name
        return name_text.strip()

    except Exception as e:
        print(f"Error during name extraction: {e}")
        return None

# Example usage:
extracted_name=extract_name(r'pan2.png')
print(extracted_name)
```
